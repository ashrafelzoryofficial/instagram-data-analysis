## Data Visualization
In this section, I use charts to understand the relationships between the variables.

### Question 6

```python
# first, load needed modals (matplotlib.pyplot ,seaborn ,plotly.express)
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px

# Then, Do more likes increase impressions?
sns.scatterplot(data=df,x=df['Likes'],y=df['Impressions'])
plt.title('Likes vs Impressions')
plt.show()
```

```python
sns.regplot(data=df,x=df['Likes'],y=df['Impressions'])
plt.title('Likes vs Impressions')
plt.show()
```

- There is a positive relationship between likes and impressions

### Question 7

```python
# Do more likes increase follows?
px.scatter(data_frame=df, x='Likes', y='Follows', trendline='ols',title='Likes vs Follows',trendline_color_override='red')
```

- Posts with more likes usually have more Follows

### Question 8

```python
# Do more hashtag impressions increase total impressions?
sns.scatterplot(data=df, x='From Hashtags', y='Impressions')
plt.title('From Hashtags vs Impressions')
plt.show()
```

```python
sns.regplot(data=df, x='From Hashtags', y='Impressions')
plt.title('From Hashtags vs Impressions')
plt.show()
```

- Posts with more hashtag impressions usually have more total impressions

### Question 10

```python
# Do more profile visits increase follows?
sns.scatterplot(data=df, x='Profile Visits', y='Follows')
plt.title('Profile Visits vs Follows')
plt.show()
```

```python
sns.regplot(data=df, x='Profile Visits', y='Follows')
plt.title('Profile Visits vs Follows')
plt.show()
```

- More profile visits can lead to more follows
